```python
import os
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

print("Working directory:", os.getcwd())
print("File present:", os.path.exists("sample_-_superstore (2).xls"))
```

```python
file_name = "sample_-_superstore (2).xls"

try:
    df = pd.read_excel(file_name)
    print("Loaded OK. Shape:", df.shape)
except Exception as e:
    print("Failed to read Excel file. Error:", e)
    print("\nIf you see an error mentioning 'xlrd', install it with:")
    print("  pip install xlrd==1.2.0   # or conda install -c anaconda xlrd=1.2.0")
    print("OR open the file in Excel and Save As -> .xlsx, then use pd.read_excel('file.xlsx')\n")
    raise
```

```python
print("Columns:",df.columns.tolist())
display(df.head(10))
print("\nInfo:")
display(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python

cols = ['Product Name', 'Quantity', 'Sales']   
df_sales = df[cols].copy()


df_sales.columns = ['product', 'quantity', 'sales']


df_sales['quantity'] = pd.to_numeric(df_sales['quantity'], errors='coerce').fillna(0).astype(int)
df_sales['sales'] = pd.to_numeric(df_sales['sales'], errors='coerce').fillna(0.0)

display(df_sales.head())
```

```python
db_path = "sales_data.db"
conn = sqlite3.connect(db_path)

#DataFrame to SQLite (replaces if exists)
df_sales.to_sql("sales", conn, if_exists="replace", index=False)
print("Wrote dataframe to", db_path, "table 'sales' (rows):", len(df_sales))

conn.close()
```

```python
conn = sqlite3.connect(db_path)

query = """
SELECT 
    product,
    SUM(quantity) AS total_qty,
    SUM(sales) AS revenue
FROM sales
GROUP BY product
ORDER BY revenue DESC
"""

df_summary = pd.read_sql_query(query, conn)
conn.close()

print("Summary (products sorted by revenue):")
display(df_summary.head(20))
```

```python
import matplotlib.pyplot as plt
top10 = df_summary.sort_values('revenue', ascending=False).head(10)

plt.figure(figsize=(12,6))  

plt.barh(top10['product'], top10['revenue'], color='skyblue')

plt.xlabel("Revenue")
plt.ylabel("Product")
plt.title("Top 10 Products by Revenue")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()

plt.savefig("top10_revenue_products.png", dpi=150, bbox_inches='tight')
```
